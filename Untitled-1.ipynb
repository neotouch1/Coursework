```python
import numpy as np
from matplotlib.image import imread
import matplotlib.pyplot as plt
import os
from FFT import fft1d, ifft1d, fft2d, ifft2d

plt.rcParams['figure.figsize'] = [5, 5]
plt.rcParams.update({'font.size': 18})
```

```python
#  
A = imread(os.path.join('170046-nochnoj_neonovyj_avtomobil-legkovyye_avtomobili-noch-neon-grafika-1920x1080.jpg'))
B = np.mean(A, -1)

#     
def pad_to_power_of_2(matrix):
    rows, cols = matrix.shape
    new_rows = 1 << (rows - 1).bit_length()
    new_cols = 1 << (cols - 1).bit_length()
    padded_matrix = np.zeros((new_rows, new_cols))
    padded_matrix[:rows, :cols] = matrix
    return padded_matrix, rows, cols

B_padded, orig_rows, orig_cols = pad_to_power_of_2(B)

plt.figure()
plt.imshow(A)
plt.axis('off')
```

```python
Bt = fft2d(B_padded)
Btsort = np.sort(np.abs(Bt.reshape(-1)))  #     
print(Btsort.shape)
```

```python
#     
for keep in (0.1, 0.05, 0.01, 0.002):
    thresh = Btsort[int(np.floor((1 - keep) * len(Btsort)))]
    ind = np.abs(Bt) > thresh  #        
    Btlow = Bt * ind
    Alow_padded = ifft2d(Btlow).real  #      FFT
    Alow = Alow_padded[:orig_rows, :orig_cols]  #    

    plt.figure()
    plt.imshow(Alow, cmap='gray')
    plt.axis('off')
    plt.title('Compressing image: ' + str(keep * 100) + '%')

plt.show()
```
